```python
from pprint import pprint

from protein_detective.uniprot import Query, search4af, search4pdb, search4uniprot, search4emdb
```

```python
import logging

# logging.basicConfig(level=logging.DEBUG)
```

```python
query = Query(
    taxon_id="9606",
    reviewed=True,
    subcellular_location_uniprot="nucleus",
    subcellular_location_go="GO:0005634",  # Cellular component - Nucleus
    molecular_function_go="GO:0003677",  # Molecular function - DNA binding
)
```

```python


results = search4uniprot(query, limit=200)
pprint(results)
```

```python


results = search4pdb(query, limit=200)
pprint(results)
```

```python
afresults = search4af(query, limit=200)
pprint(afresults)
```

```python


results = search4emdb(query, limit=200)
pprint(results)
```
